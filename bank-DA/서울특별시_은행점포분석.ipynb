```python
## 필요한 라이브러리 import
# 데이터 분석
import pandas as pd
import numpy as np

# 시각화
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
PATH_2010 ="data/서울시금융기관구별통계_2010_2014.xls"
PATH_2015 = "data/서울시금융기관구별통계_2015_2020.xls"
```

```python
bank_1 = pd.read_excel(PATH_2010)
bank_1
```

```python
# 컬럼 바꿔주기
bank_1.columns = list(bank_1.loc[1])

# 숫자만 남기고 drop
bank_1 = bank_1.loc[2:, :]

# reset index
bank_1 = bank_1.reset_index(drop = True)

# check
bank_1
```

```python
bank_2 = pd.read_excel(PATH_2015)
bank_2
```

```python
# 컬럼 바꿔주기
bank_2.columns = list(bank_2.loc[1])

# 숫자만 남기고 drop
bank_2 = bank_2.loc[2:, :]

# reset index
bank_2 = bank_2.reset_index(drop = True)

# check
bank_2
```

```python
# concat 전 컬럼의 개수가 달라서 확인
print(f"2010 ~ 2014년 데이터 shape : {bank_1.shape}")
print(f"2015 ~ 2020년 데이터 shape : {bank_2.shape}")
print("-"*100)
print(f"2010 ~ 2014년 데이터 컬럼 : {bank_1.columns}")
print(f"2015 ~ 2020년 데이터 컬럼 : {bank_2.columns}")
print("-"*100)
print(f"다른 컬럼(과거) : {set(bank_1.columns) - set(bank_2.columns)}")
print(f"다른 컬럼(미래) : {set(bank_2.columns) - set(bank_1.columns)}")
```

```python
'''
컬럼이름 맞추기(은행이름 맞추기)
* 과거 데이터
    - 한국외환은행 + 하나은행 (합쳐서) -> KEB하나은행으로 합병해서 바꿔주기
    - 수출입은행 -> 한국수출입은행으로 변경하기
    - 한국산업은행 -> KDB산업은행으로 이름 변경하기
    - 농협중앙회 -> NH농협은행으로 변경(농협중앙회의 신용사업이 NH농협금융지주로 분리되면서 자회사로 출범)

* 현재 데이터
    - 수협은행 -> 수협중앙회로 변경(수협중앙회가 수협은행 지분 100% 가짐)
'''
```

```python
# 한국외환은행 + 하나은행 (합쳐서) -> KEB하나은행
bank_1["KEB하나은행"] = bank_1['한국외환은행'].astype(np.int32) + bank_1['하나은행'].astype(np.int32)

# 한국외한은행, 하나은행 drop
bank_1 = bank_1.drop(columns = ['한국외환은행', '하나은행'])

# 수출입은행 -> 한국수출입은행 | 한국산업은행 -> KDB산업은행 | 농협중앙회 -> NH농협은행으로 변경하기
bank_1 = bank_1.rename(columns = {'수출입은행' : '한국수출입은행', '한국산업은행' : 'KDB산업은행',
                            '농협중앙회' : 'NH농협은행'})
```

```python
# 수협은행 -> 수협중앙회로 변경
bank_2 = bank_2.rename(columns = {'수협은행' : '수협중앙회'})
```

```python
# column 전처리 이후 재 확인
print(f"2010 ~ 2014년 데이터 shape : {bank_1.shape}")
print(f"2015 ~ 2020년 데이터 shape : {bank_2.shape}")
print("-"*100)
print(f"2010 ~ 2014년 데이터 컬럼 : {bank_1.columns}")
print(f"2015 ~ 2020년 데이터 컬럼 : {bank_2.columns}")
print("-"*100)
print(f"다른 컬럼(과거) : {set(bank_1.columns) - set(bank_2.columns)}")
print(f"다른 컬럼(미래) : {set(bank_2.columns) - set(bank_1.columns)}")
```

```python
# 두 데이터 병합(상·하 연결 - concat)
bank = pd.concat([bank_1, bank_2], ignore_index=True) # row bind : axis = 0, default
bank
```

```python
print(f"합친 데이터 컬럼 : {bank.columns}")
```

```python
# 결측값 확인 : 없음
bank.isnull().sum()
```

```python
# 데이터 프레임 정보 조회
bank.info()
```

```python
bank
```

```python
# 합계 컬럼은 따로 저장하기
con = (bank['자치구'] == '합계')
bank_total = bank.loc[con]

# index 초기화
bank_total = bank_total.reset_index(drop = True)
```

```python
# 자치구 컬럼에서 '합계'컬럼 drop해주기
con = (bank['자치구'] != '합계')
bank = bank.loc[con]
```

```python
# '-' 문자열 -> '0'으로 바꿔주기
bank = bank.replace('-', '0')
```

```python
# 자치구 컬럼을 제외하고 int 데이터 타입으로 변경
for col in bank.columns:
    if col != '자치구':
        bank[col] = bank[col].astype(np.int32)
```
